```python
from langchain_community.llms import Ollama

llm = Ollama(
    # model="gemma2:latest" # 사용하려는 모델이름
    # model="gemma2:2b",
    model="llama3.1:8b", # gemma2:2b, llama3.1:8b 중에서 선택하여 사용한다.
    temperature=0
)

llm.invoke("대한민국의 수도가 어디니?") # 질문
```
